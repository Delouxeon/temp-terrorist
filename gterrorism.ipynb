```python
#importing python libaries for data manipulation
import pandas as pd
import numpy as np
import matplotlib as plt 
```

```python
#Reading file into the variable "data" 
data = pd.read_csv("terrorist.txt", sep='\t')
data 
```

```python
#converting the 'features' column to string to enable splitting
fColumn = data['features'].astype(str).str.split(',')
#splitting the 'features' column into different columns to enable summation 
featureValues = pd.DataFrame(fColumn.to_list()).fillna(0).add_prefix('Feature_')
featureValues.astype(int)
```

```python
#calculating the sum of each rows
featureValues["sumColumn"] = featureValues.sum(axis =1)
featSum = featureValues["sumColumn"]
```
